# 04-03 로지스틱 손실 함수를 경사 하강법에 적용
- 분류의 정확도는 미분 가능한 함수가 아님 (상수임)
- 그렇기에 이진 크로스 엔트로피(binary cross entropy) 또는 로지스틱(logistic) 손실함수를 사용한다.
- $L = -(y\log{(a)} + (1-y)\log{(1-a))}$
  - y = 1 (양성 클래스인 경우) -> $L = -\log{(a)}$
  - y = 0 (음성 클래스인 경우) -> $L = -\log{(1-a)}$
- $w_{grad} = -(y-a)x_i$
- $b_{grad} = -(y-a)$
