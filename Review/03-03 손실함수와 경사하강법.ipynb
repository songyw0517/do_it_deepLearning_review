# 3-03 손실함수와 경사 하강법의 관계

## 손실함수
- 예측한 값과 실제 타깃값의 차이를 측정
- 회귀, 분류 등에 사용하는 손실함수가 존재한다.

## 경사하강법
- 손실함수의 차이를 줄이는 방법

# 손실함수의 종류
- ## 제곱오차
  - SE = $(y-\hat{y})^2$
  - 제곱을 하는 이유 : 오차의 크기를 최대화 또는 최소화 하기 위함
    - ### 제곱오차 손실 함수의 기울기 미분 값(그레이디언트) 
      = $-(y-\hat{y})x$
    - ### 제곱오차 손실 함수의 절편 미분 값
      = 1
    - 결론
      ```python
      err = y_i - y_hat
      w_rate = x_i
      w = w + w_rate * err
      b = b + 1 * err
      ```
