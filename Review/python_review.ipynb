# numpy

```python
# 넘파이를 이용한 배열 생성
# 읽기 편하게 된다.
# 넘파이에서 제공하는 메소드 사용가능해짐
my_arr = np.array([[10,20,30], [40,50,60]])
print(my_arr)
```

### numpy 
- 배열 타입 = ndarray

```python
# 타입은 ndarray임
print('타입은 = ',type(my_arr))
```

```python
# 넘파이 배열에서 요소 선택하기
print(my_arr[0][2])
```

### numpy.sum과 sum 비교

```python
# 넘파이 sum함수 = 배열의 모든 값을 합한 값
print('np.sum(my_arr) = ', np.sum(my_arr))

# 파이썬 내장함수의 sum함수 = 배열의 열을 합침
print('sum(my_arr) = ', sum(my_arr))
```

# 맷플롯립(matplotlib)

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4, 5],   # x 축
        [1, 4, 9, 16, 25])  # y 축
plt.show()
```

### 맷플롯립 산점도
- 각 포인트의 점만 찍는 함수
- scatter(x, y)

```python
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3, 4, 5],   # x 축
        [1, 4, 9, 16, 25])  # y 축
plt.show()
```

# numpy 배열로 산점도 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
# 표준 정규 분포를 따르는 난수 1000개 생성
x = np.random.randn(1000)

# 표준 정규 분포를 따르는 난수 1000개 생성
y = np.random.randn(1000)
plt.scatter(x, y)
plt.show()
```
