## 이진 분류
  - ## 퍼셉트론
    - 1957년에 발표, 임의의 샘플 데이터를 True 또는 False로 구분
    1. $z = \sum\limits_{i=1}^nw_i*x_i + b$
    2. z를 계단 함수에 넣음 -> y_hat (1또는 -1)
    3. y_hat을 가지고 역방향 계산 수행
    4. 반복
  - ## 아달린
    - 1960년에 발표, 퍼셉트론을 개선,
    - 적응형 선형 뉴런
    - 퍼셉트론의 경우 y_hat으로 역방향 계산을 수행했으나
    - 아달린은 z를 가지고 역방향 계산을 수행
  - ## 로지스틱 회귀
    - 1. $z = \sum\limits_{i=1}^nw_i*x_i + b$
    - 2. z를 활성화 함수(시그모이드 함수)에 넣음 -> a
    - 3. a를 가지고 역방향 계산하고
    - 4. a를 임계 함수에 넣음 -> $\hat{y}$

# 시그모이드 함수 (활성화 함수)
- 변천과정 : 오즈 비 -> 로짓 함수 -> 시그모이드 함수

  ### 오즈 비
  - $\frac{p}{1-p}$ (p는 성공 확률)
  - 확률에 따라 $0$ ~ $\infty$의 값을 가짐
  <br><br>

  ### 로짓 함수
  - 오즈 비에 로그 함수를 씌운 함수
  - $\log(\frac{p}{1-p})$ (p는 성공 확률)
  - 확률에 따라 $-\infty$ ~ $\infty$의 값을 가짐
  <br><br>

  ### 로지스틱 함수 (시그모이드 함수)
  - 로짓 함수를 확률 p에 대해 정리 (가로축을 z로 하기 위함)
  - $\log(\frac{p}{1-p}) = z$
  - $\frac{p}{1-p} = e^z$
  - $p(1+e^z) = e^z$ 
  - $p = \frac{e^z}{1+e^z} = \frac{1}{1+e^{-z}}$
